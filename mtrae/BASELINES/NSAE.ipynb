```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
from torch.utils.data.dataset import Dataset as dataset
from torch.utils.data import DataLoader
```

```python
class Dataset(dataset):
    def __init__(self, inputs, outs):
        super(Dataset, self).__init__()
        self.inputs = inputs
        self.outs = outs

    def __getitem__(self, index):
        input = self.inputs[index]
        output = self.outs[index]
        return input, output

    def __len__(self):
        return len(self.inputs)

    pass
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
class NSAE_model(nn.Module):
    def __init__(self, D_in, H=20, H2=14, latent_dim=15, classes=2, lamda1=1.0, lamda2=1.0, lamda3=1.0):
        super(NSAE_model, self).__init__()
        # self.feature = model_func()
        self.linear1 = nn.Linear(D_in, H)  # 29 * 20
        self.lin_bn1 = nn.BatchNorm1d(num_features=H)
        self.linear2 = nn.Linear(H, H2)  # 20 * 10
        self.lin_bn2 = nn.BatchNorm1d(num_features=H2)
        self.linear3 = nn.Linear(H2, H2)  # 10 * 10
        self.lin_bn3 = nn.BatchNorm1d(num_features=H2)

        self.num_class = classes
        self.fc1 = nn.Linear(H2, latent_dim)  # 10 * 7

        self.fc2 = nn.Linear(latent_dim, latent_dim-2)
        self.classifier = nn.Linear(latent_dim-2, self.num_class)  # 7 * 3

        self.fc3 = nn.Linear(latent_dim, latent_dim)  # 7 * 7
        #         self.fc_bn3 = nn.BatchNorm1d(latent_dim)
        self.fc4 = nn.Linear(latent_dim, H2)  # 7 * 10
        #         self.fc_bn4 = nn.BatchNorm1d(H2)

        self.linear4 = nn.Linear(H2, H2)  # 10 * 10
        self.lin_bn4 = nn.BatchNorm1d(num_features=H2)
        self.linear5 = nn.Linear(H2, H)  # 10 * 20
        self.lin_bn5 = nn.BatchNorm1d(num_features=H)
        self.linear6 = nn.Linear(H, D_in)  # 20 * 29
        self.lin_bn6 = nn.BatchNorm1d(num_features=D_in)
        self.relu = nn.ReLU()
        self.loss_fn = nn.CrossEntropyLoss()
        self.lamda1 = lamda1
        self.lamda2 = lamda2
        self.lamda3 = lamda3

    def encode(self, x):
        lin1 = self.relu(self.lin_bn1(self.linear1(x)))  # 29 * 20
        lin2 = self.relu(self.lin_bn2(self.linear2(lin1)))  # 20 * 10
        lin3 = self.relu(self.lin_bn3(self.linear3(lin2)))  # 10 * 10

        fc1 = self.relu(self.fc1(lin3))  # 10 * 7
        return fc1

    def decode(self, z):
        fc3 = self.relu(self.fc3(z))  # 7 * 7
        fc4 = self.relu(self.fc4(fc3))  # .view(128, -1) # 7 * 10

        lin4 = self.relu(self.lin_bn4(self.linear4(fc4)))  # 10 * 10
        lin5 = self.relu(self.lin_bn5(self.linear5(lin4)))  # 10 * 20
        return self.lin_bn6(self.linear6(lin5))  # 20 * 29

    def forward(self, x):
        feature = self.encode(x)
        x_hat = self.decode(feature)
        x_hat_feats = self.encode(x_hat)
        scores = self.classifier(self.relu(self.fc2(feature)))
        x_hat_scores = self.classifier(self.relu(self.fc2(x_hat_feats)))

        return scores, x_hat, x_hat_scores, feature

    def forward_loss(self, x, y):
        scores, x_hat, x_hat_scores, _ = self.forward(x)
        loss_c = self.loss_fn(scores, y)
        loss_c_hat = self.loss_fn(x_hat_scores, y)
        loss_mse = torch.mean((x - x_hat) ** 2)
        print(loss_c, loss_mse, loss_c_hat)
        loss = (self.lamda1 * loss_c) + (self.lamda2 * loss_mse) + (self.lamda3 * loss_c_hat)
        _, predicted = torch.max(scores.data, 1)
        correct = predicted.eq(y.data).cpu().sum()
        return loss

    def train_autoencoder(self, epoch, train_loader, optimizer):
        print_freq = 10
        avg_loss = 0
        losses = []
        for i, (x, y) in enumerate(train_loader):
            x = x.float().to(device)
            y = y.long().to(device)
            optimizer.zero_grad()
            loss = self.forward_loss(x, y)
            losses.append(loss)
            loss.backward()
            optimizer.step()
        avg_loss = torch.mean(torch.stack(losses))
        print('Epoch {:d} | Loss {:f}'.format(epoch, avg_loss))
```

```python
def train(base_loader, model, optimization, start_epoch, stop_epoch):
    model.train()
    for epoch in range(start_epoch, stop_epoch):
        model.train_autoencoder(epoch, base_loader, optimizer)

    return model
```
